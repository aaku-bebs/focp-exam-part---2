```python
strings = []

while True:
    user_input = input("Enter a string (press Enter to finish): ")
    if user_input == "":
        break
    strings.append(user_input)
```

```python
strings = [s.lower() for s in strings]
```

```python
strings.sort()
```

```python
print("\nStrings in alphabetical order:")
for s in strings:
    print(s)
```
