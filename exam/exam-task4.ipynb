```python
def contains_coconut(words):
    """
    Returns True if 'coconut' is in the list of strings, otherwise False.
    """
    return "coconut" in words
```

```python
test_list_1 = ["apple", "banana", "coconut", "date"]
test_list_2 = ["apple", "banana", "date"]
```

```python
print(contains_coconut(test_list_1))  
print(contains_coconut(test_list_2))
```
